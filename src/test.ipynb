```python
base_path = Path("../data/")

wer_results = {}

for language in ["en", "de", "hu"]:
    file_name = language + "_transcription_results.csv"
    results = pd.read_csv(base_path / file_name)
    wer = results.loc[:, "w_l_v3_tb_wer"].mean()
    wer_results[language] = wer

    if language == "en":
        wer_turbo = results.loc[:, "w_l_v3_turbo_tl_wer"].mean()
        wer_results["en_turbo"] = wer_turbo

wer_results
```
